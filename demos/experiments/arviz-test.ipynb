Test experimental arviz functionality.

```python
import pangolin
from pangolin import interface as pi
mu = pi.normal(0,1)
sigma = pi.lognormal(0,1)
y = pi.vmap(pi.normal,None,3)(mu,sigma)
dataset = pangolin.blackjax.sample_arviz(dict(mu=mu, sigma=sigma, y=y))
```

```python
dataset
```

```python
import arviz as az
az.plot_posterior(dataset);
```
