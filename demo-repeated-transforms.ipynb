```python
%load_ext autoreload
%autoreload 2
import pangolin
from pangolin import transforms
from pangolin.interface import *
from pangolin import new_infer as infer
```

```python
print("define model")
a = 2.2
b = 3.3
c = 4.4
d = 5.5
θ = normal(a,b)
z = normal(θ,c)
x = normal(z,d)
```

```python
print('original model')
display(viz_upstream([θ,z,x],[x],{θ:'θ',z:'z',x:'x'}))
```

```python
print("one round of simplification")
[θ2, z2, x2] = transforms.simplify([θ,z,x],[x],1)
display(viz_upstream([θ2,x2,z2],[x2],{θ2:'θ',z2:'z',x2:'x'}))
```

```python
print("5 rounds of simplification")
[θ2, z2, x2] = transforms.simplify([θ,z,x],[x],5)
display(viz_upstream([θ2,x2,z2],[x2],{θ2:'θ',z2:'z',x2:'x'}))
```

```python
print("13 rounds of simplification")
[θ2, z2, x2] = transforms.simplify([θ,z,x],[x],13)
display(viz_upstream([θ2,x2,z2],[x2],{θ2:'θ',z2:'z',x2:'x'}))
```

```python
print("14 rounds of simplification")
[θ2, z2, x2] = transforms.simplify([θ,z,x],[x],14)
display(viz_upstream([θ2,x2,z2],[x2],{θ2:'θ',z2:'z',x2:'x'}))
```
